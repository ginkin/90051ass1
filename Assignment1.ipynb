```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
def preprocessing():
    with open('train.txt', 'r') as f:
        train = f.read().split("\n")
        if train[-1] == "":
            train.pop(-1)
        train = [i.split(" ") for i in train]
    max_author = 0
    for i in train:
        for j in i:
            if int(j) > max_author:
                max_author = int(j)
    mat = np.zeros(shape = (max_author+1, max_author+1))
    for row in train:
        for i in range(len(row)):
            for j in range(i+1,len(row)):
                mat[int(row[i])][int(row[j])] += 1
                mat[int(row[j])][int(row[i])] += 1
    x,y = np.where(mat>0)
    edges = list(zip(list(x),list(y),list(mat[x,y].astype(int))))
    return edges
edges = preprocessing()
edges
```
